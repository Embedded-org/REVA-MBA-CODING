```
string="who is the PM of india? you know it!!"
stop_words=['who', 'is', 'the','of', 'you', 'know', 'it!!','?']
refined_str=[]
for word in string.split():
  if word.lower() in stop_words:
    pass
  else:
    refined_str.append(word)
print(refined_str)
```
