```
#TEXT CLUSTERING WITH K-MEANS
import sklearn
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.metrics import adjusted_rand_score
#we are using TfidVectorizer because we need to convert unstructured data into valid formats(structured Data)
```

```
#create document list
documents={"google plus","google mail","google translate app",
           "if you open 100 tab in google you get a smiley face.",
           "Best Cat photo","climbing ninja cat."
           "impressed with google map feedback."
           "key promoter extension for google chrome."}
    
```

```
#transform the data
vectorizer=TfidfVectorizer(stop_words='english')
X=vectorizer.fit_transform(documents)
print(X)
#here TfidVectorizer stores the data in cross matrix format.here we can see TFIDVectorizer mapped total 21 variables.
```

```
documents={"google plus",
           "google mail",
           "google translate ",
           "cat yamaguchi",
           "Best Cat photo",
           "climbing ninja cat."
           "impressed with cat"
           "key promoter extension for google chrome."}
```

```
#transform the data
vectorizer=TfidfVectorizer(stop_words='english')
X=vectorizer.fit_transform(documents)
print(X)
#here TfidVectorizer stores the data in cross matrix format.here we can see TFIDVectorizer mapped total 14 variables.
```

```
#Build the clusters
true_k=3
model=KMeans(n_clusters=true_k,init='k-means++',max_iter=100,n_init=1)
model.fit(X)
```

```
#PROFILE THE CLUSTERS
print("top terms per cluster:")
order_centroids=model.cluster_centers_.argsort()[:,::-1]
terms=vectorizer.get_feature_names()
for i in range(true_k):
  print("Cluster %d:" % i),
  for ind in order_centroids[i,:10]:
    print('%s'%terms[ind]),
  print
```

```
#use model for prediction
print("Prediction")
Y=vectorizer.transform(["google to open."])
prediction=model.predict(Y)
prediction
```
