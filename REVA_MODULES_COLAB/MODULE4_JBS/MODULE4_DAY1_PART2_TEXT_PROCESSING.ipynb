```
def wordfreq(words):
  '''function to generate the frequency distribution of the 
  given text'''
  word_freq={}
  for tok in words:
    if tok.lower() in word_freq:
      word_freq[tok.lower()]+=1
    else:
      word_freq[tok.lower()]=1
  return(word_freq)
```

```
words=soup.get_text().split()
```

```
#LEXICON BASED CLASSIFICATION
s1="i love India"
s2="i hate cigarattes"
lexicon={'love':5,'hate':-5}
lexicon['love']
```

```
lexicon['hate']
```

```
token1=s1.lower().split()
token1
```

```
strength=0
for token in token1:
  if(token in lexicon):
    value=lexicon[token]
    strength+=value
    print(strength)
  else:
    pass
if(strength>0):
  print("positive")
elif(strength<0):
  print("negative")
else:
  print("neutral")
```
