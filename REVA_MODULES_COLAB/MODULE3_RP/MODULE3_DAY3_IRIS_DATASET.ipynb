```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
plt.show()
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = "all"
```

```
import seaborn
import sklearn
seaborn.get_dataset_names()
```

```
iris=sns.load_dataset('iris')
iris.head()
```

```
iris.describe()
```

```
np.round(iris.mean(),2)
```

```
np.round(iris.median(),2)
```

```
iris.count()
```

```
#counting frequency for categorical variables
pd.crosstab(index=iris['species'],columns="frequency")
```

```
correlation=iris.corr()
correlation
```

```
sns.heatmap(correlation,cmap="RdYlGn",annot=True)
plt.show()
```

```
sns.heatmap(correlation,cmap="Blues",annot=True)
plt.show()
```

```
#compute percentile
np.round(iris.quantile((np.arange(0,1,0.05))),2)  #from 0% to 100% in the gap of 5%
```

```
#pairplot
sns.pairplot(iris,hue="species",palette="RdYlGn",size=5,diag_kind='hist')
sns.pairplot?
plt.show()
```

```
sns.pairplot(iris,hue="species",palette="copper",size=5,x_vars=["sepal_width","sepal_length"],y_vars=["petal_width","petal_length"])
```

```
sns.pairplot(iris,kind="reg")
plt.show()
```

```
plt.figure(figsize=(15,10))
plt.subplots_adjust(hspace=0.2,wspace=0.2)
plt.subplot(2,2,1)
plt.title("sepal_length")
plt.plot(iris.sepal_length,'r')
plt.subplot(2,2,2)
plt.title("sepal_width")
plt.plot(iris.sepal_width,'b')
plt.subplot(2,2,3)
plt.title("petal_length")
plt.plot(iris.petal_length,'g')
plt.subplot(2,2,4)
plt.title("petal_width")
plt.plot(iris.petal_width,'y')
plt.show()
```

```
#histogram and kernel density estimator
plt.figure(figsize=(15,10))
sns.distplot(iris.sepal_width)
plt.show()
```

```
#BOX PLOT
plt.figure(figsize=(15,10))
sns.boxplot(data=iris,orient="v",palette="Set2")
plt.show()
```

```
#BOX PLOT
plt.figure(figsize=(15,10))
sns.boxplot(x='species',y='sepal_length',data=iris,linewidth=2)
plt.show()
#BOX PLOT
plt.figure(figsize=(15,10))
sns.boxplot(x='species',y='petal_length',data=iris,linewidth=2)
plt.show()
```
