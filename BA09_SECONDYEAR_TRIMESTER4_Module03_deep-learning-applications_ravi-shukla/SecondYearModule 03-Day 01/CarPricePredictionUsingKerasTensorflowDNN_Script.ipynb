```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from google.colab import drive
drive.mount('/content/drive')
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab_Notebooks/SECONDYEAR_TRIMESTER4_Module03_deep-learning-applications/DAY1_APPLICATION_OF_DEEP_LEARNING/CarDetailsV3.csv')
df.head(2)
```

```python
df['mileage'] = df.mileage.str.split(" ",expand = True)[0]
df['engine'] = df.engine.str.split(" ",expand = True)[0]
df['max_power'] = df.max_power.str.split(" ",expand = True)[0]
df.dtypes
df.head(5)
```

```python
df['mileage'] = df['mileage'].astype(float)
df['engine'] = df['engine'].astype(float)
df['seats'] = df['seats'].astype(float)
df2 = df.drop(['name','torque','max_power'],axis='columns')
df2['year'] = 2023 - df2['year']
df2.head(5)
```

```python
df2['mileage']=df2['mileage'].fillna(df2['mileage'].mean())
df2['engine']=df2['engine'].fillna(df2['engine'].mean())
df2['seats']=df2['seats'].fillna(df2['seats'].mean())
df2.info()
```

```python
df2.describe()
```

```python
data = pd.get_dummies(df2)
data.head(5)
```

```python
from mlxtend.preprocessing import minmax_scaling
data1=minmax_scaling(data, columns=['selling_price', 'km_driven'])
data = data.drop(['selling_price','km_driven'],axis='columns')
```

```python
data['selling_price']=data1['selling_price']
data['km_driven']=data1['km_driven']
data.head(5)
```

```python
from sklearn.model_selection import train_test_split
data.columns
x = data.drop('selling_price',axis=1)
y = data['selling_price']
```

```python
x.info()
```

```python
train_data, test_data, train_label,test_label = train_test_split(x,y, test_size=0.2)
len(train_data)

from sklearn import preprocessing
train_data = preprocessing.normalize(train_data)
test_data = preprocessing.normalize(test_data)
```

```python
import tensorflow as tf
models=tf.keras.models
layers=tf.keras.layers
model = models.Sequential()
model.add(layers.Dense(512,activation='relu',input_shape=(train_data.shape[1],)))
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dense(64,activation='relu'))
model.add(layers.Dense(1))
model.compile(optimizer='rmsprop',loss='mse',metrics=['mae'])
model.fit(train_data,train_label,epochs=10)
```

```python
test_results = model.evaluate(test_data, test_label, verbose=0)
test_results
```
