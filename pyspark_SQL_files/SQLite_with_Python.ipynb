```
#Setting up the cloud environment
#The below code set up necessary libraries so that SQL database file (here, flight_information.db) 
#can be imported from Google drive directly.
# Installing the PyDrive in Python Environment in Colab
!pip install -U -q PyDrive

# Importing the necessary libraries prior Authentcation
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
from google.colab import auth
from oauth2client.client import GoogleCredentials
```

```
#Next task is Authentication of the user 
auth.authenticate_user()
gauth = GoogleAuth()
gauth.credentials = GoogleCredentials.get_application_default()
drive = GoogleDrive(gauth)
```

```
#Finally, the flight database will loaded into the Colab local environment from Google drive. 
#For this, the database file has been shared with public access. 
#Steps to follow

#Assign the variable link to the google drive link (as copied below) of the file.
#Next, separate the id value from the link part to get the content of the file.
link = 'https://drive.google.com/open?id=1Z1PTRj0pdf4tUSu5RZqj2doFYn37W8Ce'
split, id = link.split('=')
downloaded = drive.CreateFile({'id':id}) 
downloaded.GetContentFile('flight_information.db')
```

```
#Settting up the SQLite connection and displaying the first 2 rows from the table airlines
#Steps to follow
#Import the SQLite3 module (which is pre-installed in Google colab)
#Execute SQL command to display the first couple of rows from the airlines table
import sqlite3
conn = sqlite3.connect("flight_information.db")
cur = conn.cursor()
list(cur.execute("select * from airlines limit 2;"))
```
